```python
import pandas as pd
import numpy as np
import random

# Set random seed for reproducibility
np.random.seed(42)

# Simulated categories
channels = ['Facebook', 'Instagram', 'Google Ads', 'TikTok', 'Email']
age_groups = ['18-24', '25-34', '35-44', '45-54', '55+']
genders = ['Male', 'Female', 'All']
campaign_names = [f"FashionPromo_{i}" for i in range(1, 21)]

# Function to simulate a single row of campaign data
def generate_campaign_row():
    campaign = random.choice(campaign_names)
    channel = random.choice(channels)
    age = random.choice(age_groups)
    gender = random.choice(genders)
    
    budget = round(np.random.uniform(500, 5000), 2)
    impressions = int(budget * np.random.uniform(10, 30))
    ctr = round(np.random.uniform(0.01, 0.12), 4)
    clicks = int(impressions * ctr)
    conversion_rate = round(np.random.uniform(0.005, 0.10), 4)
    conversions = int(clicks * conversion_rate)
    revenue = round(conversions * np.random.uniform(30, 150), 2)

    return {
        "Campaign Name": campaign,
        "Channel": channel,
        "Target Age Group": age,
        "Target Gender": gender,
        "Budget ($)": budget,
        "Impressions": impressions,
        "Clicks": clicks,
        "CTR (%)": round(ctr * 100, 2),
        "Conversions": conversions,
        "Conversion Rate (%)": round(conversion_rate * 100, 2),
        "Revenue ($)": revenue
    }

# Generate 1000 rows
data = [generate_campaign_row() for _ in range(1000)]
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("marketing_campaign_data.csv", index=False)
print("✅ Dataset generated and saved as 'marketing_campaign_data.csv'")
```
